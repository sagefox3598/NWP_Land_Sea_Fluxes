```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr
from IPython.display import display, HTML

# Define the file paths and river codes
sim_file = '../../../data/streamflow/vic_flow/q_sim_PMET.csv'
obs_file = '../../../data/streamflow/gauge_flow/q_obs_PMET.csv'
river_codes = ['X11342001', 'X11147001', 'X11040001', 'X10523002', 'X10704002']

# Define the mapping from river codes to river names
river_names = {
    'X11342001': 'Rio Aysen en Puerto Aysen (Aysen)',
    'X11147001': 'Rio Cisnes en Puerto Cisnes (Cisnes)',
    'X11040001': 'Rio Palena Bajo Junta Rosselot (Palena)',
    'X10523002': 'Rio Puelo en Carrera Basilio (Puelo)',
    'X10704002': 'Rio Futaleufu ante Junta Rio Malito (Yelcho)'
}

# Load the data
sim_df = pd.read_csv(sim_file, parse_dates=['Date'])
obs_df = pd.read_csv(obs_file, parse_dates=['Date'])

# Ensure the Date column is the index for easy merging
sim_df.set_index('Date', inplace=True)
obs_df.set_index('Date', inplace=True)

# Join the dataframes on the Date index
merged_df = sim_df[river_codes].join(obs_df[river_codes], lsuffix='_sim', rsuffix='_obs', how='inner')

# Initialize a list to store statistical results
stats_results = []
annual_mean_results = []
monthly_mean_error_results = []

# Function to calculate statistical metrics
def calculate_statistics(sim, obs, river_name):
    valid_idx = ~obs.isna()
    sim = sim[valid_idx]
    obs = obs[valid_idx]

    if len(sim) == 0 or len(obs) == 0:
        return None

    pearson_corr, _ = pearsonr(sim, obs)
    spearman_corr, _ = spearmanr(sim, obs)
    bias = np.mean(sim - obs)
    nash_sutcliffe = 1 - (np.sum((sim - obs)**2) / np.sum((obs - np.mean(obs))**2))

    return {
        'River Name': river_name,
        'Pearson Correlation': pearson_corr,
        'Spearman Correlation': spearman_corr,
        'Bias': bias,
        'Nash-Sutcliffe Efficiency': nash_sutcliffe
    }

# Plot and calculate statistics for each river
for river_code in river_codes:
    river_name = river_names[river_code]
    combined_df = merged_df[[f'{river_code}_sim', f'{river_code}_obs']].dropna()

    if not combined_df.empty:
        plt.figure(figsize=(12, 6))
        plt.plot(combined_df.index, combined_df[f'{river_code}_sim'], label='Simulated', color='blue')
        plt.plot(combined_df.index, combined_df[f'{river_code}_obs'], label='Observed', color='orange', alpha=0.7)
        plt.title(f'Streamflow Comparison for {river_name}')
        plt.xlabel('Date')
        plt.ylabel('Streamflow')
        plt.legend()
        plt.grid(True)
        plt.show()

        stats = calculate_statistics(combined_df[f'{river_code}_sim'], combined_df[f'{river_code}_obs'], river_name)
        if stats:
            stats_results.append(stats)

        # Plot 7-day rolling averages
        plt.figure(figsize=(12, 6))
        plt.plot(combined_df.index, combined_df[f'{river_code}_sim'].rolling(window=7).mean(), label='Simulated 7-Day Rolling Avg', color='blue')
        plt.plot(combined_df.index, combined_df[f'{river_code}_obs'].rolling(window=7).mean(), label='Observed 7-Day Rolling Avg', color='orange', alpha=0.7)
        plt.title(f'7-Day Rolling Average Streamflow for {river_name}')
        plt.xlabel('Date')
        plt.ylabel('Streamflow')
        plt.legend()
        plt.grid(True)
        plt.show()

        # Plot 30-day rolling averages
        plt.figure(figsize=(12, 6))
        plt.plot(combined_df.index, combined_df[f'{river_code}_sim'].rolling(window=30).mean(), label='Simulated 30-Day Rolling Avg', color='blue')
        plt.plot(combined_df.index, combined_df[f'{river_code}_obs'].rolling(window=30).mean(), label='Observed 30-Day Rolling Avg', color='orange', alpha=0.7)
        plt.title(f'30-Day Rolling Average Streamflow for {river_name}')
        plt.xlabel('Date')
        plt.ylabel('Streamflow')
        plt.legend()
        plt.grid(True)
        plt.show()

        # Calculate mean monthly streamflow and standard deviation, averaged over all years
        monthly_mean_sim = combined_df[f'{river_code}_sim'].groupby(combined_df.index.month).mean()
        monthly_mean_obs = combined_df[f'{river_code}_obs'].groupby(combined_df.index.month).mean()
        monthly_std_sim = combined_df[f'{river_code}_sim'].groupby(combined_df.index.month).std()
        monthly_std_obs = combined_df[f'{river_code}_obs'].groupby(combined_df.index.month).std()

        # Plot mean monthly streamflow with error bars
        months = range(1, 13)
        plt.figure(figsize=(12, 6))
        plt.bar(np.array(months) - 0.2, monthly_mean_sim, width=0.4, yerr=monthly_std_sim, label='Simulated', color='blue', alpha=0.7)
        plt.bar(np.array(months) + 0.2, monthly_mean_obs, width=0.4, yerr=monthly_std_obs, label='Observed', color='orange', alpha=0.7)
        plt.title(f'Mean Monthly Streamflow with Error Bars for {river_name}')
        plt.xlabel('Month')
        plt.ylabel('Mean Streamflow')
        plt.xticks(months, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
        plt.legend()
        plt.grid(True)
        plt.show()

        # Calculate mean annual streamflow
        annual_mean_sim = combined_df[f'{river_code}_sim'].resample('Y').mean()
        annual_mean_obs = combined_df[f'{river_code}_obs'].resample('Y').mean()

        mean_annual_sim = annual_mean_sim.mean()
        mean_annual_obs = annual_mean_obs.mean()
        percent_diff = ((mean_annual_sim - mean_annual_obs) / mean_annual_obs) * 100
        mae = np.mean(np.abs(annual_mean_sim - annual_mean_obs))

        annual_mean_results.append({
            'River Name': river_name,
            'Mean Annual Simulated Streamflow': mean_annual_sim,
            'Mean Annual Observed Streamflow': mean_annual_obs,
            '% Difference': percent_diff,
            'Mean Absolute Error': mae
        })

        # Calculate monthly mean error
        monthly_error = combined_df[f'{river_code}_sim'] - combined_df[f'{river_code}_obs']
        monthly_mean_error = monthly_error.groupby(combined_df.index.month).mean()
        monthly_mean_error_results.append({
            'River Name': river_name,
            **{f'Mean Error Month {month}': error for month, error in monthly_mean_error.items()}
        })

# Split data into first and second halves for analysis
split_date = merged_df.index[len(merged_df) // 2]

first_half_df = merged_df[merged_df.index <= split_date]
second_half_df = merged_df[merged_df.index > split_date]

def calculate_half_statistics(df, half):
    results = []
    for river_code in river_codes:
        river_name = river_names[river_code]
        combined_df = df[[f'{river_code}_sim', f'{river_code}_obs']].dropna()
        if not combined_df.empty:
            annual_mean_sim = combined_df[f'{river_code}_sim'].resample('Y').mean()
            annual_mean_obs = combined_df[f'{river_code}_obs'].resample('Y').mean()

            mean_annual_sim = annual_mean_sim.mean()
            mean_annual_obs = annual_mean_obs.mean()
            percent_diff = ((mean_annual_sim - mean_annual_obs) / mean_annual_obs) * 100
            mae = np.mean(np.abs(annual_mean_sim - annual_mean_obs))

            results.append({
                'River Name': river_name,
                'Mean Annual Simulated Streamflow': mean_annual_sim,
                'Mean Annual Observed Streamflow': mean_annual_obs,
                '% Difference': percent_diff,
                'Mean Absolute Error': mae
            })
    return results

first_half_results = calculate_half_statistics(first_half_df, 'First Half')
second_half_results = calculate_half_statistics(second_half_df, 'Second Half')

# Convert results to DataFrames and display
stats_df = pd.DataFrame(stats_results)
annual_mean_df = pd.DataFrame(annual_mean_results)
first_half_df = pd.DataFrame(first_half_results)
second_half_df = pd.DataFrame(second_half_results)
monthly_mean_error_df = pd.DataFrame(monthly_mean_error_results)

# Display the DataFrames in Jupyter Notebook with titles
display(stats_df)
display(annual_mean_df)
display(HTML('<h2>First Half Statistics</h2>'))
display(first_half_df)
display(HTML('<h2>Second Half Statistics</h2>'))
display(second_half_df)
display(monthly_mean_error_df)
```
